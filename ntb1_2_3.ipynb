```python
# Data Frame with out NaN Colunms
dfnc = df.drop(["created_at_month","rent_to_single","rent_type","transformed_credit","transformed_rent","has_security_guard",
                "property_type","regular_person_capacity","has_business_deed","extra_person_capacity","cost_per_extra_person",
                "rent_price_on_regular_days","rent_price_on_special_days","rent_price_at_weekends","location_radius"],axis=1)
```

```python
# Duplicated rows
dfnc.duplicated(keep=('first'))
```

```python
# Delete duplicated rows
dfnc.drop_duplicates(inplace=True,keep='first')
```

```python
# Reset indexes
dfnc.reset_index(inplace=True , drop=True)
dfnc
```

```python
# Shape of new data frame
dfnc.shape
```

```python
# Focse on sell and house and delete the offices , presells and rents
dfnc = dfnc[~dfnc["cat2_slug"].str.contains("rent",na=True)]
dfnc = dfnc[~dfnc["cat3_slug"].str.contains("rent","office",na=True)]
dfnc = dfnc[~dfnc["cat3_slug"].str.contains("presell",na=True)]
dfnc.shape
```
